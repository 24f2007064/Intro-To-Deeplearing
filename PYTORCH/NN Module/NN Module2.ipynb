## Question

Build a feed-forward neural network with the following architecture:

- Input Layer: **5 input features**
- Hidden Layer: **3 neurons** with **ReLU** activation
- Output Layer: **1 neuron** with **Sigmoid** activation

```text
                 Hidden Layer (ReLU)
 Input Layer        3 Neurons             Output Layer (Sigmoid)

x1 ● ───────────────┐
x2 ● ───────────┐   │
x3 ● ────────┐  │   │      ● h1 ───┐
x4 ● ─────┐  │  │   ├────► ● h2 ───┼────► ● y_pred
x5 ● ──┐  │  │  │   │      ● h3 ───┘
       └──┴──┴──┴───┘

Input → Hidden:
Weights = 5 × 3
Biases  = 3

Hidden → Output:
Weights = 3 × 1
Biases  = 1
```

````python
import torch
import torch.nn as nn
````

````python
class Model(nn.Module):
    def __init__(self, num_input_features):

        super().__init__()


        # INPUT LAYER of 5 to--> 3
        self.linear1 = nn.Linear(num_input_features, out_features = 3 )

        # Hidden LAYER
        self.relu = nn.ReLU()

        # another layer of 3 to--> 1
        self.linear2 = nn.Linear(3,1)

        #output laye
        self.sigmoid = nn.Sigmoid()

    def forward (self, features):
        out = self.linear1(features)
        out = self.relu(out)
        out = self.linear2(out)
        out = self.sigmoid(out)

        return out
````

#### Creating Synthetic Dataset

````python
features = torch.rand(10,5)
features.shape
````

````python
model = Model(features.shape[1])

````

````python
model(features)
````

````python
class con_Model(nn.Module) :

    def __init__(self, num_input_features):
        super().__init__()

        self.network = nn.Sequentials(
            nn.Linear(num_input_features,3),
            nn.ReLU(),
            nn.Linear(3,1),
            nn.Sigmoid()
        )

    def forward(self, features):
        out = self.network(features)
        return out
        
````
