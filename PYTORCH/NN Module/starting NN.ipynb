#### REbuilding that pytorc Training PIPELINE with NN module

```python
import torch
import torch.nn as nn
```

```python
class model(nn.Module):

    def __init__(self, num_features):


        # Initialize the parent nn.Module class so this model gets all PyTorch module features (parameters, training, saving, etc.)
        super().__init__()

        # sinngle leyer nuron
        self.linear = nn.Linear(num_features, out_features= 1)
        self.sigmoid = nn.Sigmoid()

    def forward (self, features):
        out = self.linear(features)              # we are doing similer like z = WX + B
        out = self.sigmoid(out)                  # we are dping y = sigmoid(z)
        return out
```

#### Creating Synthetic Dataset

```python
features = torch.rand(10,5)
features.shape[1]
```

#### Creating Model

```python
model = model(features.shape[1])


# FORWARD PASS

#model.forward(features)    ----> we dont need to do this as NN module recomend to do model(features) directly
model(features)             #-----> as we had 10 rows
```


**`model(features)`** passes the input to the model. PyTorch automatically calls the `forward()` method internally, so we don't need to write `model.forward(features)` explicitly.

```python
# SHOW MODEL's WEIGHT and BIAS

print(model.linear.bias)
model.linear.weight
```
