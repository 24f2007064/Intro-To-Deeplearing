```python
import torch
```

```python
a = torch.tensor([1,2,3])
b = a

a,b
```

a and b point to the same memory location.
They are not two separate tensors.

Therefore, if we change a,
b will also change automatically.


example-->

```python
a[1] = 99
a
```

```python
b
```

#### To create an independent copy, use clone().

```python
v = torch.tensor([1,2,3])
v
```

```python
g = v.clone()
g
```

```python
v[1] = 99
v
```

```python
g
```

#### clone() creates a new tensor with its own memory location.
#### Changes in one tensor will not affect the other.
