```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace= True)
df.shape
```

##### Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:], df.iloc[:, 0])

X_train.head()
```

```python
y_train.head()
```

#### Scaling

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train, X_test
```

#### Lablel Encoder

```python
label = LabelEncoder()
y_train = label.fit_transform(y_train)
y_test = label.transform(y_test)

y_test
```

#### Numpy array to Torch

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
bias = torch.rand(1, dtype=torch.float64, requires_grad= True)
bias
```

#### Defing Model

```python
class simpleNN():

    #Dfining the weights and Bias
    def __init__(self, x):
        self.weight = torch.rand(x.shape[1], 1, dtype= torch.float64, requires_grad= True )
        self.bias = torch.rand(1, dtype=torch.float64, requires_grad= True)

    # Making the FORWARD Relationships
    def forward(self, x):
        z = torch.matmul(x, self.weight) + self.bias
        y_preds = torch.sigmoid(z)
        return y_preds
    
    # Making the LOSS Functions
    def loss_fun (self, y_preed, y):
        epsilon = 1e-7
        y_preed = torch.clamp(y_preed, epsilon, 1-epsilon)

        loss = -(y*torch.log(y_preed) + (1 - y)*torch.log(1-y_preed)).mean()
        return loss 
```

```python
learning_rate = 0.1
epochs = 45
```

#### Training Pipeline

```python
model = simpleNN(X_train_tensor)

for epoch in range(epochs):
    # forward Pass
    y_pred = model.forward(X_train_tensor)


    # loss calculate
    loss = model.loss_fun(y_pred, y_train_tensor)
    print (f"{epoch} and loss {loss.item()}")

    # Backward pass
    loss.backward()
    
    # parametes Update
    with torch.no_grad():
        model.weight -= learning_rate * model.weight.grad       # we are doing GRADINT DECENT ---> Wnew = Wold - learingrate(dL/dW)
        model.bias -= learning_rate * model.bias.grad
    
    # Zero Grads
    model.weight.grad.zero_()
    model.bias.grad.zero_()
 
```

```python
# Model Evaluation
with torch.no_grad():                     # Disable gradient calculation

    y_pred = model.forward(X_test_tensor) # Predict probabilities

    y_pred = (y_pred > 0.5).float()       # Convert probabilities to 0 or 1

    accuracy = (y_pred == y_test_tensor).float().mean()  # Calculate accuracy

    print(f"Accuracy: {accuracy.item()}") # Print accuracy
```
