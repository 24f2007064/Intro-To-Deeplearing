### REDUCTIONS Operation

```python
e = torch.randint(size = (3,2), low = 1, high = 6)
e
```

```python
#sum
torch.sum(e)
```

```python
# sum of columns -->0
torch.sum(e, dim = 0)
```

```python
# Sum of ROWs -->1
torch.sum(e, dim =1)
```

```python
f = torch.randint(size = (3,2), low = 1, high = 6, dtype= torch.float32)
f
```

```python
torch.mean(f)
```

```python
torch.mean(f, dim = 0)
```

```python
torch.median(f, dim = 1)
```

similarly we have MAX, MEAN STD, PRODUCT, VAR

```python
torch.argmax(f)   # Position of max element
```

```python
torch.argmin(f)
```
