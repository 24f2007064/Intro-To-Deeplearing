#### PERMUTE

Changes the order of dimensions (axes) of a tensor.
It does not change the values, only how they are organized.
Useful when data is stored in one format but a model expects another.

```python
y.permute(1,2,0)
```

```python
y.permute(1,2,0).size()
```

### UNSQUIZEE


##### Adds a new dimension of size 1 at a specified position.
##### Commonly used to create batch dimensions for neural networks.

```python
c = torch.rand(226,226,3)       #image size in geenral
c.shape
```

```python
c.unsqueeze(1).shape
```

# SQUEZEE

#####  Removes dimensions whose size is 1.

```python
v.squeeze().shape
```
