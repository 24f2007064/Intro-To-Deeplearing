### QUEST 1, Simple NN

```python
# Question:
# Implement Logistic Regression for a single training example using PyTorch tensors.
#
# Given:
# x = 8.11  (CGPA)
# y = 1     (Placed)
#
# Initialize:
# w = 0.5
# b = -3.0
#
# Tasks:
# 1. Compute the linear transformation:
#       z = w*x + b
#
# 2. Apply the sigmoid activation:
#       y_hat = 1 / (1 + exp(-z))
#
# 3. Compute the Binary Cross Entropy Loss:
#       L = -(y*log(y_hat) + (1-y)*log(1-y_hat))
#
# 4. Enable gradient tracking for w and b.
#
# 5. Perform backpropagation using:
#       loss.backward()
#
# 6. Print:
#       z
#       y_hat
#       loss
#       w.grad
#       b.grad
#
# 7. Update the parameters using Gradient Descent:
#       w = w - lr*w.grad
#       b = b - lr*b.grad
#
# Use learning rate:
#       lr = 0.01
```

```python
x = torch.tensor([6.13])
y = torch.tensor([0])
```

```python
w = torch.tensor(1.0, requires_grad= True)
b = torch.tensor(0.0, requires_grad= True)
w,b
```

```python
z = w*x + b
y_pred = torch.sigmoid(z)
z, y_pred
```

```python
def binary_cross_entropy_loss(c,d):
    L = -(d*torch.log(c) + (1-d)*torch.log(1-c))
    return L

loss = binary_cross_entropy_loss(y_pred, y)

loss
```

```python
loss.backward()
```

```python
w.grad, b.grad
```
